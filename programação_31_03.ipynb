```python
#Estrutura de decicões

#Escrever um programa que recebe o nome e a idade do usuário
#Verifica se o usuário é criança,adolecente e adulto

# >= 12 anos é criança
# =< 17 anos é adolescente
# adulto

#Entrada

nome = str (input("digite o seu nome: "))
idade = int (input("digite a sua idade: "))

#Processamento / Saida

if idade <= 12:
  print(f"{nome}, voce tem {idade} anos de idade,vc é crianca.")

elif idade <= 18:
   print(f"{nome} voce tem {idade} anos de idade,voce é adolescente.")

else:
  print(f"{nome},voce tem {idade} anos de idade, voce é adulto.")

```

```python
#Em um jogo,um oersonagem possui uma variavel chamada energia.O programa deve:
#Se a energia for maior que 70 - Forte,se a energia estiver entre 30 e 70 - Cansado,
#Se a energia estiver menor que 30 - Fraco.

#Entrada

energia = int(input("digite a energia do seu personagem: "))

#Processamento / saida

if energia <= 30:
  print(f"seu personagem tem energia igual a {energia}.Ele está fraco.")

elif energia <= 70:
   print(f"seu personagem tem energia igual a {energia}.Ele está cansado.")

else:
  print(f"seu personagem tem energia igual a {energia}.Ele está forte.")

```

```python
#Escrever um programa que classifica os tipos de triangulos e valida se é um triangulo.

#Validar se os lados digitados pode ser formar um triangulo.
#Lado + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1

#Tipos de triangulos
#Equilatero - todos os lados iguais - ==
#Escaleno - todos os lados diferentes - !=
#Isoceles - dois lados iguais

#Entrada
lado1 = int(input("digite os lado do triangulos: "))
lado2 = int(input("digite os lado do triangulos: "))
lado3 = int(input("digite os lado do triangulos: "))

#Verificar se é um triangulo válido

if lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1:
    print("triangulo válido")
    if lado1 == lado2 and lado1 == lado3:
      print("triangulo equilatero")
    elif lado1 != lado2 and lado1 != lado3 and lado2 != lado3:
      print("triangulo escaleno")
    else:
      print("triangulo isoceles")
else:
    print(f"não é um triangulo inválido")
```
